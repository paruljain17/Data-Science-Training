```python
#translation --- changes the location of an image on the screen
import numpy as np
rows,col= img.shape[:2]
tx ,ty=1000,500 #translation factor
M = np.float32([[1,0,tx],[0,1,ty]])
#[1,0,tx]
##1 -- keep the x coordimates same
##0 -- dont change teh y coordinates
#tx ---- translation factor

translated_image=cv2.warpAffine(img,M,(col,rows))
#wrapAffine= isse apn kafi sare functions perform krwa skte like scaling, translation ,rotation ,shearing and many more
plt.imshow(translated_image)
```

```python
#rotation
center= (col//2,rows//2) #center of the image
angle=45
scale=0.5
#scale ---zoom in and zooom out
#1 ---original image ,<1=zoom out,>1=zoom in

M=cv2.getRotationMatrix2D(center,angle,scale)
rotated_image=cv2.warpAffine(img,M,(col,rows))
plt.imshow(rotated_image)
```

```python
#flipping -- mirror image
flip_image =cv2.flip(img,0)
#flip code=0 --- flip vertically
#flip code=1 --- flip horizontally
# flip code = -1 --- flip both

plt.imshow(flip_image)
```
