```python
#to change our income column data type to int 
df[' income'] = df[' income'].map({' <=50K': 0, ' >50K': 1})
df.head()
```

```python
cat_col=df.select_dtypes(include="object")
cat_col
```

```python
num_col=df.select_dtypes(exclude="object")
num_col
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_col_encoded = cat_col.apply(le.fit_transform)
cat_col_encoded
```

```python
final_df = pd.concat([cat_col_encoded, num_col], axis=1)
```

```python
##x and y 
x = final_df.drop(" income",axis=1)
y = final_df[" income"]
```

```python
##train test split 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, 
                                                    y,
                                                      test_size=0.2,
                                                        random_state=67)
```

```python
#standard scaler == it is used to normalize the data == means in range
#it make mean=0  and std=1
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
#svc
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train_scaled,y_train)
```

```python
svc.score(x_test_scaled,y_test)
```

```python
svc.score(x_train_scaled,y_train)
```

```python
svm = SVC()
svm_cv = GridSearchCV(estimator=svm, param_grid=param_grid, cv=5)
svm_cv.fit(x_train_scaled, y_train)
```
