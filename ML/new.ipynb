```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cust_satisfaction.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df["Customer Type"].value_counts()
```

```python
loyal_customer = df[df["Customer Type"]=="Loyal Customer"]
disloyal_customer = df[df["Customer Type"]=="disloyal Customer"]
```

```python
loyal_customer = loyal_customer.sample(20000)
loyal_customer.shape #shape define the rows and columns
```

```python
balance_df = pd.concat([loyal_customer,disloyal_customer],axis=0)
balance_df.shape
```

```python
cat_col =balance_df.select_dtypes(include=["object"])
cat_col.head()
```

```python
num_col =balance_df.select_dtypes(exclude=["object"])
num_col.head()
```

```python
#pd.get_dummies(cat_col,drop_first=True).astype(int).head(10)
```

```python
cat_col['Customer Type'].value_counts()
```

```python
dt ={
  "Loyal Customer":0,
    "disloyal Customer":1 
}
```

```python
cat_col["Customer Type"]=cat_col["Customer Type"].map(dt)
```

```python
cat_col.head(20)
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

ohe = OneHotEncoder(drop="if_binary")
cat_col_encoded = ohe.fit_transform(cat_col).toarray()
cat_col_encoded
```

```python
column_name = list(ohe.get_feature_names_out())
column_name
```

```python
one_hot = pd.DataFrame(cat_col_encoded,columns=column_name)
one_hot.head()
```
