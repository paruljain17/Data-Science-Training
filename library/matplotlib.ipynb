```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
arr = np.random.randint(20000,40000,100)
arr
```

```python
#lineplot
#it works on numerical data
arr.sort()
plt.plot(arr)
plt.show()
```

```python
x= [1,2,3,4,5]
y =[1,10,9,10,25]
plt.plot(x,y,marker="s",ms="10",mfc="red",mec="yellow",color="black",
         lw=5,ls="dotted",label="lineplot")
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
#ms=marker size
#mfc=marker face colour
#mec=marker edge colur
```

```python
#scatterplot -- numerical data
plt.scatter(x,y)
plt.title("scatterplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#bargraph-->numerical and categorical
x = ["mp","up","rj","gj"]
y =[10,20,30,40]
colours =["orange","green","blue","yellow"]
hatch_patterns = ["//","\\\\","xx","oo"]
#plt.bar(x,y,color=colours,hatch=hatch_patterns,
        #width=0.4,edgecolor="black",alpha=0.6,linewidth=3)
        
plt.barh(x,y)
plt.grid()
#plt.legend()
plt.show()
```

```python
#histogram ---distribution of data 
arr = np.random.randint(20000,40000,100)
plt.hist(arr,bins=20)
plt.show()
```

```python
#piechart --->percentage
x =["mp","up","rj","gj"]
y =[10,20,30,40]
explodes = (0.3,0.1,0,0)
plt.pie(y,labels=x,autopct="%1.2f%%",shadow=True, explode=explodes )
plt.savefig("piechart.pdf")
plt.show()
```

```python
#subplot
import warnings
warnings.filterwarnings("ignore")
x = np.linspace(0,10,100)
plt.figure(figsize=(10,6))
plt.subplot(2,2,1) #2 rows ,2 columns , 1st graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2) #2 rows ,2 columns , 2nd graph
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3) #2 rows ,2 columns , 3rd graph
plt.plot(x,np.tan(x))
plt.title("tan")
plt.subplot(2,2,4) #2 rows ,2 columns , 4th graph
plt.plot(x,np.log(x))
plt.title("log")
plt.tight_layout()
plt.show()
```
