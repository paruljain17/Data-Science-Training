```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#its a preloaded data set
df = sns.load_dataset("tips")
df.head()
```

```python
sns.lineplot(df["total_bill"])
```

```python
sns.scatterplot(df["total_bill"])
```

```python
sns.scatterplot(x="total_bill",y ="tip",data=df)
plt.title("scatterplot")
plt.show()
```

```python
sns.histplot(df["total_bill"],color="orange",edgecolor="black",lw=4)
plt.show()
```

```python
sns.barplot(x="day",y="size",data=df,lw=5)
```

```python
sns.kdeplot(df["total_bill"])
```

```python
#sns --- 1 categorical --- catplot
#2 numerical --- scatter plot ,lineplot --- relational---relplot
```

```python
#if we dont define kind then by default it show scatter plot 
sns.relplot(x="total_bill",y="tip",data=df,kind="scatter",
            hue="day",style="smoker",
            row="day")
plt.show()
```

```python
#swarm avoid the overlapping
#catplot
sns.catplot(x="day",y="total_bill",data=df,kind="swarm")
```

```python
#counplot
sns.countplot(x="day",data=df)
```

```python
#barplot
sns.barplot(x="day",y="total_bill",data=df,hue="sex")
```

```python
#pointplot --- shows the mean of the data connected by lines 
sns.pointplot(x="day",y="total_bill",data=df)
```

```python
#boxplot ----spread of data in the form of box
#circle=outliers(exception)
sns.boxplot(x="day",y="total_bill",data=df)
```

```python
#outliers handle
q1 = df["total_bill"].quantile(0.25)
q2 = df["total_bill"].quantile(0.75)
iqr =q2-q1
lower_limit = q1-1.5*iqr
upper_limit = q2+1.5*iqr
print(lower_limit,upper_limit)
df[df["total_bill"]>upper_limit]
df = df[(df["total_bill"]>lower_limit)&(df["total_bill"]<upper_limit)]
sns.boxplot(df["total_bill"])

```

```python
#violinplot -- spread of data in the form of violin
sns.violinplot(x="day",y="total_bill",data=df)
```

```python
sns.pairplot(df)
```

```python
sns.histplot(df["tip"],kde=True)
```
